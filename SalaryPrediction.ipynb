```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score


data = pd.read_csv('dataset.csv')
data.head()

x = data.drop(columns=['Salary'])
y = data["Salary"]

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3)

# Xây dựng model
reg_model = LinearRegression()
reg_model.fit(x_train.values, y_train)

# Dự đoán
predictions = reg_model.predict(x_test)
# print(predictions)
# print(y_test)

# Tính toán độ chính xác
score = r2_score(predictions, y_test)
score
```

```python
plt.scatter(x_train, y_train, color='red', )
plt.plot(x_train,reg_model.predict(x_train), color='blue')
plt.title("Linear Regression salary vs experinces - train", size=18)
plt.xlabel("Year of Employee",size=15)
plt.ylabel("Saleries of Employee", size=15)
plt.show()
```

```python
plt.scatter(x_test, y_test, color='red', )
plt.plot(x_test,reg_model.predict(x_test), color='blue')
plt.title("Linear Regression salary vs experinces - test", size=18)
plt.xlabel("Year of Employee",size=15)
plt.ylabel("Saleries of Employee", size=15)
plt.show()
```

```python
# dự đoán lương bằng cách nhập từ bàn phím
ex_prediction = reg_model.predict([[2],[6],[5]])
ex_prediction
```
